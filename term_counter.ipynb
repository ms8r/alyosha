Gets number of search results for a list of search terms across different sources and compares relative frequencies.

Note: For the match counting part use termcntr.py instead

```python
from alyosha import alyosha as al
from alyosha import reference as REF
import time
from random import random
import numpy as np
import pandas as pd
import re
import logging
import json
```

```python
base_term = 'political'
test_terms = ['"super rich"', '"rising inequality"', '"scale back government"', 
    '"affordable care"', '"gay movement"', '"gay rights"', '"scientific evidence"',
    '"same sex marriage"', '"gay marriage"', '"whistleblower edward snowden"',
    '"religious liberty"', '"religious tolerance"', '"religious beliefs"',
    '"religious convictions"', '"efficient market"', '"free market"',
    '"government intervention"', '"american values"', '"single mother"',
    '"single parent"', '"national security"', '"minimum wage"', '"tree hugger"', 
    '"endangered species"', '"austerity measures"', 
    '"fiscal discipline"', '"feminist movement"', '"christian values"',
    '"the feds"', '"federal government"', '"environmental activist"', 'discrimination'
]
max_sleep = 10
sources = [s.site for s in sorted(REF.source_sites, key=lambda k: k.specVal, reverse=True)]
```

```python
data = np.zeros((len(sources), len(test_terms) + 1), np.int)
result = pd.DataFrame(data, index=sources, columns=[base_term] + test_terms)
gs = al.GoogleSerp()
for src in sources:
    time.sleep(random() * max_sleep)
    try:
        result.ix[src, base_term] = gs.search(base_term, exact=True, site=src)
        logging.info("%s:%s:%d", src, base_term, result.ix[src, base_term])
    except al.EmptySearchResult:
        result.ix[src, base_term] = 0
        logging.info("%s:%s:%d", src, base_term, result.ix[src, base_term])
    except (al.ResultParsingError, al.PageRetrievalError) as e:
        logging.warning("%s:%s: %s", src, base_term, e)
    for t in test_terms:
        time.sleep(random() * max_sleep)
        try:
            result.ix[src, t] = gs.search(t, exact=True, site=src)
            logging.info("%s:%s:%d", src, t, result.ix[src, t])
        except al.EmptySearchResult:
            result.ix[src, t] = 0
            logging.info("%s:%s:%d", src, t, result.ix[src, t])
        except al.ResultParsingError:
            logging.warning("%s:%s: ResultParsingError", src, t)
        except al.PageRetrievalError:
            logging.warning("%s:%s: PageRetrievalError", src, t)
            
with open('term_counts.json', 'w') as fout:
    result.to_json(fout)
```
