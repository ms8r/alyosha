```python
import os
from itertools import combinations_with_replacement
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from  matplotlib.font_manager import FontProperties
from alyosha import reference as REF
```

```python
sources = sorted([(s.site, s.specVal) for s in REF.source_sites], key=lambda k: k[1])
sources = [s for s, t in sources]
```

```python
with open('termcntr.done', 'r') as fin:
    recs = [tuple(r.rstrip().split('\t')) for r in fin.readlines()]
```

```python
recs[:3]
```

```python
terms = list(set([t for s, t, c in recs]))
```

```python
v = np.zeros((len(sources), len(short_terms)))
v = np.nan
tdf = pd.DataFrame(v, index=sources, columns=terms)  
```

```python
for s, t, c in recs:
    tdf[t][s] = int(c)
```

```python
tdf.head()
```

```python
ntdf = pd.DataFrame(v, index=sources, columns=terms)
```

```python
ntdf = ntdf.drop('weeklystandard.com', axis=0)
```

```python
for t in terms:
    ntdf[t + '_norm'] = tdf[t] / tdf.political.map(float)
    ntdf[t + '_norm'] = ntdf[t + '_norm'] / float(ntdf[t + '_norm'].max())
ntdf = ntdf.drop(terms, axis=1)
```

```python
ntdf.head()
```

```python
save_path = 'term_plots'
pos = np.arange(len(sources) - 1) + 0.5   # dropped weeklystandard
for t in [s + '_norm' for s in terms]:
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_yticks(pos + 0.5)
    ax.set_yticklabels(sources)
    ax.set_title(t)
    ax.barh(pos, ntdf[t].values)
    plt.savefig(os.path.join(save_path, t.replace('"', '').replace(' ', '_') + '.png'), bbox_inches='tight')
    plt.close()
```

```python
short_terms = [
        '"super rich"',
        '"rising inequality"',
        '"scientific evidence"',
        '"same sex marriage"',
        '"efficient market"',
        '"free market"',
        '"government intervention"',
        '"american values"',
        '"austerity measures"',
        '"fiscal discipline"',
        '"christian values"',
        'discrimination',
]
```

```python
kill_terms = [k + '_norm' for k in (set(terms) - set(short_terms))]
```

```python
cntdf = ntdf.drop(kill_terms, axis=1)
```

```python
cntdf.columns = cntdf.columns.map(lambda k: k.replace('_norm', '').replace('"', ''))
```
